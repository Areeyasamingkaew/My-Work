```
Sales['SalesChannel'] = Sales['SalesChannel'].replace('online', 'website')
Sales # จากข้อมูลมันเป็น online อยู่ก่อนเเล้วค่ะ หนูจึงเปลี่ยนจาก online เป็น  website เเทนนะคะ เพื่อความเข้าใจค่ะ
```

#7 เปลี่ยนประเภทข้อมูลใน unit sold ให้เป็น int

8 เพิ่ม column ใหม่ 

```
#8 เพิ่มคอลัมใหม่ดังนี้
Sales['TotalRevenue'] = Sales['UnitsSold']*Sales['UnitPrice']  #or  Sales.insert(10, "TotalRevenue", (Sales.UnitsSold) * (Sales.UnitPrice))
Sales['TotalCost'] = Sales['UnitCost']*Sales['UnitPrice']      #or  Sales.insert(10, "TotalCost", (Sales.UnitCost) * (Sales.UnitPrice)) 
Sales['Profit'] = Sales['TotalRevenue']-Sales['TotalCost']     #or Sales.insert(10, "Profit", (Sales.TotalRevenue - Sales.TotalCost)) 
Sales
```
